NumPy: intended to bring performance and functionality improvements for numerical computing.


*   From User Perspective: Efficiently broadcast operations across dimensions, provide implementations of many functions across linear algebra, statistics,...
*   From Programming Perspective: enable other package to use numpy arrays as an efficient data interface, efficiently process data without type-checking overhead
*   From Hardware Perspective: enable easy file save and load of n-d arrays, efficiently store n-d arrays in vectorised form to benefit from DRAM locality




NumPy provides


1.   extension package to Python for multi-dimensional arrays
2.   closer to hardware (efficiency)
3.   designed for scientific computation (convenience)
4.   also known as array oriented computing



**Why it is useful:**: memory efficient container that provides fast numerical operations

**Copies and Views**


A slicing operation creates a view on the original array, which is just a way of accessing array data. Thus the original array is not copied in memory. You can use **np.may_share_memory()** to check if two arrays share the same memory block.


**When modifying the view, the original array is modified as well:**

```python
a = np.arange(10)
a
```

```python
b = a[::2]
b
```

```python
np.shares_memory(a, b)
```

```python
b[0] = 10
b
```

```python
a  #even though we modified b,  it updated 'a' because both shares same memory
```

```python
a = np.arange(10)
c = a[::2].copy()     #force a copy
c
```

```python
np.shares_memory(a, c)
```

```python
c[0] = 10
a
```

**Fancy Indexing**

NumPy arrays can be indexed with slices, but also with boolean or integer arrays **(masks)**. This method is called **fancy indexing**. It creates copies not views.

**Using Boolean Mask**

```python
a = np.random.randint(0, 20, 15)
a
```

```python
mask = (a % 2 == 0)
mask
```

```python
extract_from_a = a[mask]

extract_from_a
```

```python
extract_from_a[0] += 1 # original array is not modifies
print(extract_from_a)
print(a)
```

**Indexing with a mask can be very useful to assign a new value to a sub-array:**

```python
a[mask] = -1
a
```


# **Exercise on finding number of points outside n-dimensional sphere**

```python
ndim = 2
npoints = 100000
```

```python
points = np.random.rand(npoints, ndim)
```

```python
points[0:2, :]
```

```python
dfo = np.zeros((npoints, 1)) # distance from origin
outside_points = 0
```

```python
%%time

for i in range(npoints):
    for j in range(ndim):
        dfo[i] += points[i, j] ** 2
    dfo[i] = np.sqrt(dfo[i])
    if dfo[i] > 1:
        outside_points += 1
```

```python
print("Fraction of points outside is", outside_points/npoints)
```

```python
# 1 - (pi / 4)
1 - 3.14/4
```

```python
# Another approach
%%time
sq_points = points * points
dfo = np.sqrt(np.sum(sq_points, axis = 1))
outside_points = np.sum(dfo > 1) / npoints
```

```python
print("Fraction of points outside is ", outside_points/npoints)
```

```python
%%time
outside_points = np.sum(np.sqrt(np.sum(points * points, axis = 1)) > 1) / npoints
```

```python
def area_outside_square(npoints, ndim):
    points = np.random.rand(npoints, ndim)
    return np.sum(np.sqrt(np.sum(points * points, axis = 1)) > 1) / npoints
```

```python
area_outside_square(100000, 2)
```

```python
for i in range(2, 11):
    print(i, area_outside_square(1000000, i))
```
